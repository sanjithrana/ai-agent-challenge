```python
import fitz  # PyMuPDF (fitz)
import pandas as pd
from io import StringIO

def parse(pdf_path):
    """
    Extract ICICI Bank PDF data using PyMuPDF (fitz) and return a pandas DataFrame.

    Args:
        pdf_path (str): Path to the ICICI Bank PDF file.

    Returns:
        pd.DataFrame: Extracted data in a pandas DataFrame.
    """
    
    doc = fitz.open(pdf_path)

    
    text = ''
    for page in doc:
        text += page.get_text()

    rows = text.split('\n')
    data = []
    for row in rows:

        row = row.strip()

        if row:

            columns = row.split(':')
          
            if len(columns) >= 2:
                
                column_values = [column.strip() for column in columns]
                
                data.append(column_values)


    df = pd.DataFrame(data)


    return df


pdf_path ='data\\icici\\icici_sample.pdf'
df = parse(pdf_path)
print(df)
```

```python

import fitz  # PyMuPDF
import pandas as pd

def parse(pdf_path):
    """
    Extracts relevant data from an ICICI Bank PDF statement.

    Args:
        pdf_path (str): Path to the ICICI Bank PDF statement.

    Returns:
        pd.DataFrame: A pandas DataFrame containing the extracted data.
    """
    data = []
    doc = fitz.open(pdf_path)
    
    for page_num in range(len(doc)):
        page = doc[page_num]
        text = page.get_text()
        lines = text.split('\n')
        lines = [line.strip() for line in lines]
        
        account_number = None
        account_name = None
        transaction_date = None
        transaction_amount = None
        
        for line in lines:
            if 'Account Number:' in line:
                account_number = line.split(':')[-1].strip()
            elif 'Account Name:' in line:
                account_name = line.split(':')[-1].strip()
            elif 'Date:' in line:
                transaction_date = line.split(':')[-1].strip()
            elif 'Amount:' in line:
                transaction_amount = line.split(':')[-1].strip()
        
        if account_number and account_name and transaction_date and transaction_amount:
            data.append({
                'Account Number': account_number,
                'Account Name': account_name,
                'Transaction Date': transaction_date,
                'Transaction Amount': transaction_amount
            })
    
    doc.close()
    df = pd.DataFrame(data)
    return df

# Example usage
pdf_path = 'data//icici_sample.pdf'
df = parse(pdf_path)
print(df)
```
